# Tratamento dos dados

```python
import pandas as pd # type: ignore
```

```python
measurements_all_ocassions = pd.read_csv('dados/ahi-cesd.csv')
measurements_all_ocassions
```

```python
measurements_pre_intervention = measurements_all_ocassions.query('occasion == 0')
measurements_pre_intervention
```

```python
measurements_pre_intervention = measurements_pre_intervention.set_index('id')

measurements_pre_intervention = \
    measurements_pre_intervention.rename_axis(index=None)

measurements_pre_intervention
```

```python
measurements_post_intervention = measurements_all_ocassions.query('occasion == 1')
measurements_post_intervention
```

```python
measurements_post_intervention = measurements_post_intervention.set_index('id')

measurements_post_intervention = \
    measurements_post_intervention.rename_axis(index=None)

measurements_post_intervention
```

```python
measurements_post_intervention = measurements_post_intervention[['ahiTotal', 'cesdTotal']]
measurements_post_intervention
```

```python
measurements_pre_post_intervention = measurements_pre_intervention.join(measurements_post_intervention, how='inner', lsuffix='.pre', rsuffix='.post')
measurements_pre_post_intervention
```

```python
measurements_pre_post_intervention = \
    measurements_pre_post_intervention.drop(columns=[
        'occasion',
        'elapsed.days'
    ])

measurements_pre_post_intervention
```

```python
measurements_pre_post_intervention['ahiChange'] = \
    measurements_pre_post_intervention['ahiTotal.post'] - measurements_pre_post_intervention['ahiTotal.pre']

measurements_pre_post_intervention
```

```python
measurements_pre_post_intervention['ahiResponse'] = \
    measurements_pre_post_intervention['ahiChange'] >= 5

measurements_pre_post_intervention
```

```python
measurements_pre_post_intervention['cesdChange'] = \
    measurements_pre_post_intervention['cesdTotal.post'] - measurements_pre_post_intervention['cesdTotal.pre']

measurements_pre_post_intervention
```

```python
measurements_pre_post_intervention['cesdResponse'] = \
    measurements_pre_post_intervention['cesdChange'] <= -5

measurements_pre_post_intervention
```

```python
demographics = pd.read_csv('dados/participant-info.csv')
demographics
```

```python
demographics = demographics.set_index('id')

demographics = \
    demographics.rename_axis(index=None)

demographics
```

```python
demographics = demographics.drop(columns=['intervention'])
demographics
```

```python
data = demographics.join(measurements_pre_post_intervention, how='inner')
data
```

```python
data = data.query('intervention != 4')
data
```

```python
data = data.drop(columns=['intervention'])
data
```

# Separação dos dados

```python
predicting_variables = data.drop(columns=[
    'ahiTotal.post',
    'cesdTotal.post',
    'ahiChange',
    'cesdChange',
    'ahiResponse',
    'cesdResponse'
])

predicting_variables
```

```python
outcome_variable = data['cesdResponse']
outcome_variable
```

```python
import numpy as np
```

```python
np.random.seed(1234)
```

```python
from sklearn.model_selection import train_test_split # type: ignore
```

```python
train_x, test_x, train_y, test_y = train_test_split(predicting_variables, outcome_variable, test_size=0.2)

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
data['cesdResponse'].value_counts(normalize=True)
```

```python
train_y.value_counts(normalize=True)
```

```python
test_y.value_counts(normalize=True)
```

# Treinamento do modelo

```python
from sklearn.tree import DecisionTreeClassifier # type: ignore
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=5)

model.fit(train_x, train_y)
```

# Avaliação do desempenho

```python
from sklearn.metrics import accuracy_score # type: ignore
```

```python
predicted_y = model.predict(test_x)

accuracy_score(test_y, predicted_y)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

confusion_matrix_display = ConfusionMatrixDisplay.from_predictions(
    test_y,
    predicted_y,
    cmap='Blues',
    colorbar=False,
    display_labels=['NR', 'R'])

confusion_matrix_display.ax_.set(xlabel='Valor predito', ylabel='Valor real')
```

```python
import matplotlib.pyplot as plt

from sklearn.tree import plot_tree

plt.figure(figsize=(18, 18))

plot_tree(model,
          feature_names=model.feature_names_in_,
          class_names=['NR', 'R'],
          fontsize=12,
          filled=True,
          rounded=True)

plt.show()
```
